```python
from sklearn.preprocessing import LabelEncoder
labelencoder=LabelEncoder()
for column in df.columns:
    df[column] = labelencoder.fit_transform(df[column])
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings('ignore')
# "class" column as numpy array.
y = df["class"].values

# All data except "class" column.
x = df.drop(["class"], axis=1).values

# Split data for train and test.
x_train, x_test, y_train, y_test = train_test_split(x,y,random_state=42,test_size=0.2)
x_train, y_train
x_test, y_test
```

# Classification using Logistic Regression Technique

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
lr.fit(x_train,y_train)
```

```python
print("Test Accuracy: {}%".format(round(lr.score(x_test,y_test)*100,2)))
```

```python
print("Test Accuracy: {}%".format(round(lr.score(x_train,y_train)*100,2)))
```

# Classification using Support Vector Machine Technique

```python
from sklearn.svm import SVC

SVM = SVC(random_state=42, gamma="auto")
SVM.fit(x_train, y_train)
```

```python
print("Test Accuracy: {}%".format(round(SVM.score(x_test,y_test)*100,2)))
```

```python
print("Test Accuracy: {}%".format(round(SVM.score(x_train,y_train)*100,2)))
```

# Classification using K- Nearest Neighbors Technique

```python
from sklearn.neighbors import KNeighborsClassifier as KNC

bestK_value = 0
bestK_score = 0

for i in range(1,10):
    knn = KNC(n_neighbors=i)
    knn.fit(x_train, y_train)
    if knn.score(x_test, y_test)>bestK_score:
        bestK_score = knn.score(x_train, y_train)
        bestK_value = i
```

```python
print("""Best KNN Value: {}
Test Accuracy: {}%""".format(bestK_value,round(knn.score(x_test,y_test)*100,2)))
```

# Classification using Naive Bayes Technique

```python
from sklearn.naive_bayes import GaussianNB

NB = GaussianNB()
NB.fit(x_train,y_train)
```

```python
print("Test accuracy: {}%".format(round(NB.score(x_test, y_test)*100,2)))
```

# Classification using Decision Tree Technique

```python
from sklearn.tree import DecisionTreeClassifier

DTC = DecisionTreeClassifier()
DTC.fit(x_train, y_train)
```

```python
print("Test Accuracy: {}%".format(round(DTC.score(x_test, y_test)*100,2)))
```

# Random Forest Classifier
# In this step, I will train a Random Forest Classifier model and check the results.

```python
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
rf = RandomForestClassifier(oob_score=True)
rf.fit(trainX, trainY)

train_scoreRF = rf.score(trainX, trainY)
oob_score = rf.oob_score_
Adjusted_R2_trainRF = 1 - (1 - rf.score(trainX, trainY)) * (len(trainY) - 1) / (len(trainY) - trainX.shape[1] - 1)
Adjusted_R2_testRF = 1 - (1 - rf.score(testX, testY)) * (len(testY) - 1) / (len(testY) - testX.shape[1] - 1)

print('Train Adjusted R2: %', Adjusted_R2_trainRF * 100)
print('Test Adjusted R2: %', Adjusted_R2_testRF * 100)
print('OOB Score: %', oob_score * 100)
```
